```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Ace Superstore Retail Dataset.xlsx")
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
#Looking for sample data where Category Name is missing
df[df['Sub-Category'].str.contains('Cheese', case=False, na=False)]
```

```python
# Creating Dictionary for mapping Sub-Category to Category
mapping_dict = df.dropna(subset=['Category']).drop_duplicates(subset=['Sub-Category']).set_index('Sub-Category')['Category'].to_dict()
```

```python
# Fill missing Category values using Sub-Category to Category mapping
df['Category'] = df['Category'].fillna(df['Sub-Category'].map(mapping_dict))
```

```python
df.head()
```

```python
# Looking for Null Values in Category
df['Category'].isnull().sum()
```

```python
# Dropping all the null values for consistency
df = df.dropna(subset=['Category'])
```

```python
df['Category'].isnull().sum()
```

```python
# Exporting the cleaned data in a new spreadsheet
df.to_excel('Karunesh_Sehgal_Cleaned_dataset.xlsx', index=False)
```
